### **Task 1**: 
Plot the distribution of combined fuel mileage (column 'comb', in miles per gallon) by manufacturer (column 'make'), for all manufacturers with at least eighty cars in the dataset. Consider which manufacturer order will convey the most information when constructing your final plot. 

**Hint**: Completing this exercise will take multiple steps! Add additional code cells as needed in order to achieve the goal.

```python
number_of_manufactures = df['make'].value_counts()
number_of_manufactures
```

```python
filter = number_of_manufactures >= 80
dict=df.value_counts('make')[filter].to_dict()
dict
```

```python
filter2=df['make'].isin(df.value_counts('make')[filter].index)
df[filter2]
```

```python
df[filter2]['make'].map(dict)
```

```python
filteres_df = df[filter2].copy()
filteres_df
```

```python
filteres_df['manf_count'] = df[filter2]['make'].map(dict)
filteres_df.sort_values(by='manf_count',ascending=False,inplace=True)
filteres_df.head()
```

```python
grid = sb.FacetGrid(filteres_df.loc[: , ['make','comb']],col = 'make',col_wrap=3)
```

```python
grid.map_dataframe(sb.histplot,data = "comb")
grid.set_ylabels("Fuel mileage")
grid.savefig('result1.png')
```

### **Task 2**: 
Continuing on from the previous task, plot the mean fuel efficiency for each manufacturer with at least 80 cars in the dataset.  Can you use error bars?

```python
maker = df[filter2].groupby('make')
```

```python
mean = pd.DataFrame(maker['comb'].mean())
mean.sort_values(by = 'comb',ascending = False,inplace = True)
mean
```

```python
sb.barplot(x = df[filter2]['comb'],y = df[filter2]['make'],order = mean.index)
```
