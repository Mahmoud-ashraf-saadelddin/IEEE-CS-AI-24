#  Top 10 countries with Netflix Content

```python
df.groupby("country").show_id.count().sort_values(ascending=False).head(10)
```

# Monthly Trend of Movie and TV Show Releases

```python
df['date_added']=pd.to_datetime(df['date_added'])
df['month'] = df['date_added'].dt.month
df.groupby('month').release_year.count()
```

#  Yearly Trend of Movie and TV Show Releases

```python
df['year'] = df['date_added'].dt.year
df.groupby('year').release_year.count()
```

# Top 10 Movies

```python
filter = df['type'] == 'Movie'
df[filter].sort_values(by='release_year', ascending=False).head(10)
```

# Top 10 TV Show

```python
filter = df['type'] == 'TV Show'
df[filter].sort_values(by='release_year', ascending=False).head(10)
```
