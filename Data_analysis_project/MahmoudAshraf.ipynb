```python
# removing any non numeric values of 'Expected year of graduation' column
df['Expected year of graduation'] = df['Expected year of graduation'].str.replace(r'\D', 'null', regex=True)
df['Expected year of graduation'] = pd.to_numeric(df['Expected year of graduation'], errors='coerce')

df.head(50)
```

```python
df['Expected year of graduation'] = df['Expected year of graduation'].dropna()
df=df.dropna(axis='index', how='all', subset=['Expected year of graduation'])
df.tail(50)
```

```python
df['City']=df['City'].str.replace('الزقازيق', 'Zagazig')
df['City']=df['City'].str.replace('ابوكبير', 'Abo_keber')
df['City']=df['City'].str.replace('القاهره', 'Cairo')
df['City']=df['City'].str.replace('ديرب نجم', 'dirb_negm')
df['City']=df['City'].str.replace('منيا القمح', 'dirb_negm')
df['City']=df['City'].str.replace('Zagazig-الشرقية', 'Zagazig')
arabic_pattern = r'[\u0600-\u06FF]'
df = df[~df['City'].str.contains(arabic_pattern, na=False)]
df.tail(50)
```

```python
df['Faculty']=df['Faculty'].str.replace('هندسة','Engineering')
df['Faculty']=df['Faculty'].str.replace('كليه الهندسه','Engineer')
df['Faculty']=df['Faculty'].str.replace('كلية هندسة ','Engineer')
df.info()
```
